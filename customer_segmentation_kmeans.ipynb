```python
# Customer Segmentation â€“ KMeans Clustering

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# 1. Sample dataset (customers: Age vs Annual Income)
data = {
    "Age": [25, 30, 45, 35, 23, 40, 55, 60, 48, 50],
    "Annual_Income": [30000, 40000, 60000, 50000, 25000, 65000, 80000, 85000, 70000, 75000]
}
df = pd.DataFrame(data)
```

```python
# 2. Apply KMeans
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df["Cluster"] = kmeans.fit_predict(df[["Age", "Annual_Income"]])
```

```python
# 3. Visualization
plt.scatter(df["Age"], df["Annual_Income"], c=df["Cluster"], cmap="viridis", s=100)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=200, c="red", marker="X", label="Centroids")
plt.xlabel("Age")
plt.ylabel("Annual Income")
plt.title("Customer Segmentation (KMeans)")
plt.legend()
plt.show()

print(df)
```
